```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

Different graph types:

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# change from scatter to bar graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# change to a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# change to a histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) # don't need to specify x, since the frequency is given by number of y elements within each bin
```

```python
# define graph (a line plot)
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# add elements to the plot (title, axes labels, layout parameter)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

Changing text elements and labels:

```python
# import font dictionary to change font type, colour, and size later
font1 = {'family':'sans-serif',
         'color':'blue',
         'size':20}
font2 = {'family':'monospace',
         'color':'green',
         'size':14}

# define plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# add plot elements, now with specified font parameters
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# left-align the title label
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

Changing data plot elements (symbols and lines):

```python
fig, ax = plt.subplots(figsize=(5, 3))

# scatter plot with custom data point symbol and colour
ax.scatter(x, y, 
           marker='*', # NEW ADDITIONS 
           color = "indigo") # NEW ADDITIONS

#fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# line plot with custom data point symbol, colour, and line style
ax.plot(x, y, 
           marker='*', # NEW ADDITIONS 
           color = 'indigo',
           linestyle = '--',
           linewidth = 2) # NEW ADDITIONS

#fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# modify colour by hex code
ax.plot(x, y, 
           marker='*', # NEW ADDITIONS 
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2) # NEW ADDITIONS

#fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# further marker modifications (separate from line)
ax.plot(x, y, 
           marker='*', 
           markersize = 12, # NEW ADDITIONS 
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359', # NEW ADDITIONS, changing symbol outline
           markerfacecolor = '#fa9359') # NEW ADDITIONS, changing symbol fill
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y, 
           marker='*', 
           markersize = 12, 
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#fa9359', 
           markerfacecolor = '#fa9359') 

# add grid lines
#ax.grid(axis = 'y')

# add complex grid lines
ax.grid(#axis = 'y',
        color = 'red',
        linestyle = '-.',
        linewidth = 1)
```
