```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# create figure w multiple axes (ax1, ax2 --> plots)

# define layout of plots: 1 row with 2 columns
fig,(ax1, ax2) = plt.subplots(ncols = 2, 
                                nrows = 1,
                                figsize = (7,3))

# define plots
ax1.scatter(x1,y1, color='pink')

ax2.bar(x2,y2)
plt.xlabel("Name",  
           fontdict = {'family': 'sans serif', 'color': 'indigo', 'fontsize':12})

fig.show()
```

```python
# alterate subplot arrangement method

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], # all plots encapsulated in []
                                    ['ax2', 'ax3']], # each row encapsulated by groups of []
                                    figsize = (7,4),
                                    layout = "constrained") # adjusts subplots to ensure elements all fit in layout


# use for loop to put text in each subplot
#for label, ax in someaxes.items(): 
#    ax.text(0.5, 0.5, f'This is {label!r}',
#            fontsize = 14,
#            ha = 'center',
#            transform = ax.transAxes)

# populate subplots with data
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

# set axes labels for subplots
someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)

plt.show()
```

Multiple viz on one axes object:

```python
# sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define fig and axes
fig, ax = plt.subplots(figsize=(7,3))

# call both bar and plot elements to the same axes (ax)
ax.bar(x,y1,
       color = "teal")
ax.plot(x, y2, color = "red")

# add error info
y2_sd = np.std(y2) # calculate standard deviation
ax.errorbar(x,
            y2,
            yerr = y2_sd, # y error
            fmt = "none", # only plot error, not actual data points
            ecolor = "black",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery = 2 # specify data intervals for which error bars will be shown
            )

fig.show()
```

Add images to plots:

```python
# add from internet

# load libraries
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images

# get images from the internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)

# create figures and line plot
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color = "red")

# overlay container for new axis (image) on figure to act as container for image
ax_image = fig.add_axes([0.1, # x coord on fig (not axes)
                         0.11, # y coord
                         0.15, # image width
                         0.35] #image height
                         )

ax_image.imshow(image) # add image to image container
ax_image.axis('off')

ax.set_title('Power Levels of Pirate Crew', fontdict = {'family': 'sans serif', 'color': 'maroon', 'fontsize': 12})

plt.show()
```
