Customizing plots:

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# create line plot with both y variables on same axis
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1") # add label to plot obj
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right', # add legend, using loc arg to specify legend position
          frameon = True, # add frame around legend
          fontsize = 12, # set font size
          ncol = 2, # set number of columns
          shadow = True,
          bbox_to_anchor = (1,1) # move legend outside of plot, specifying location by coordinates (relativeto loc)
          ) 

fig.show()
```

Annotating plots with text()

```python
# create scatter plot showing both y variables on the same axes
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# specifiy text, its coordinates, and other text attributes
ax.text(10, 95, "This value is important!",
        ha = 'center', # text alignment
        color = 'red',
        size = 20
        )
fig.show()
```

Positioning text annotations:

```python
# create empty plot with x and y ranges of 0-10
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# position text based on axis or figure
ax.text(1,5, ". Data:(1,5)", transform = ax.transData) # move 1 unit across and 5 units up (loc by data)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform = ax.transAxes) # move 50% across, 10% up (loc by plot/axis)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform = fig.transFigure) # move 20% across and up (loc by whol figure)
```

Annotating with arrows:

```python

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10,95), xytext=(20,94),
            #arrowprops=dict(facecolor = 'black', shrink=0.05) # black arrow
            arrowprops = dict(arrowstyle = "wedge", color = "hotpink"))
```

Modifying axes:

```python
# create scatter plot w 2 y variables on same axes
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # remove everything with _locator
#ax.xaxis.set_major_formatter(plt.NullFormatter()) # remove only labels with _formatter, keeping axis ticks
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # limit max number of ticks to 3
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # set tick mark intervals to 5

# modify x axis tickmarks
plt.xticks(rotation=45, ha='right', 
           fontsize = 12)

# modify x axis title
plt.xlabel("Shiny New X Axis!",  
           fontdict = {'family': 'serif', 'color': 'indigo', 'fontsize':18})

fig.show()
```
